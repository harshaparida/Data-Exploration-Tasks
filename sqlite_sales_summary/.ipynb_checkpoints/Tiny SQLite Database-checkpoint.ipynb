# Tiny SQLite Database

In this task, I created a simple sales analysis script using Python, SQLite, pandas, and matplotlib. The goal was to connect to a local SQLite database (sales_data.db), extract basic sales metrics using SQL, and visualize them.

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('sales_data.db')
cursor = conn.cursor()
```

```python
cursor.execute('''
    CREATE TABLE IF NOT EXISTS sales (
        id INTEGER PRIMARY KEY,
        product TEXT,
        quantity INTEGER,
        price REAL
    )
''')
```

```python
sample_data = [
    ('Wireless Mouse', 120, 15.99),
    ('Mechanical Keyboard', 75, 45.50),
    ('USB-C Hub', 60, 22.75),
    ('Laptop Stand', 40, 34.99),
    ('Noise Cancelling Headphones', 30, 89.95),
    ('Webcam HD', 50, 39.90),
    ('Portable SSD 1TB', 25, 99.99),
    ('Bluetooth Speaker', 65, 55.49),
    ('Ergonomic Chair', 15, 199.99),
    ('Monitor 24-inch', 20, 129.99)
]

# Clear old data and insert fresh
cursor.execute('DELETE FROM sales')
cursor.executemany('INSERT INTO sales (product, quantity, price) VALUES (?, ?, ?)', sample_data)
conn.commit()
```

```python
query = '''
    SELECT product, SUM(quantity) AS total_quantity,
           SUM(quantity * price) AS total_revenue
    FROM sales
    GROUP BY product
'''
df = pd.read_sql_query(query, conn)
```

```python
# 5. Print output
print("=== Basic Sales Summary ===")
print(df)
```

```python
plt.figure(figsize=(12, 6))

# Quantity bar plot
plt.subplot(1, 2, 1)
plt.bar(df['product'], df['total_quantity'], color='skyblue')
plt.title('Total Quantity Sold')
plt.ylabel('Quantity')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 2)
plt.bar(df['product'], df['total_revenue'], color='lightgreen')
plt.title('Total Revenue')
plt.ylabel('Revenue ($)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
